```python
import cv2
#%matplotlib inline
from matplotlib import pyplot as plt
%pylab inline

cap = cv2.VideoCapture("rtsp://jamescheng:jamescheng@140.113.215.27:554/live.sdp")
ret,frame = cap.read()
a = 0
img_array = []
while ret and (a<50):
    ret, frame = cap.read()
    height, width, layers = frame.shape
    size = (width,height)
    img_array.append(frame)
    cv2.waitKey(5)
    a+=1
    
out = cv2.VideoWriter('out.avi',cv2.VideoWriter_fourcc(*'DIVX'), 5, size)
for i in range(len(img_array)):
    out.write(img_array[i])
out.release()
```
